```python
import sqlite3
import pandas as pd

db_path = "sqlite_db_pythonsqlite.db"
conn = sqlite3.connect(db_path)

cursor = conn.cursor()
```

Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members!

```python
query = """
SELECT
    f.name AS facility_name,
    SUM(
        CASE
            WHEN b.memid = 0 THEN f.guestcost * b.slots  -- Guest cost
            ELSE f.membercost * b.slots  -- Member cost
        END
    ) AS total_revenue
FROM Bookings b
JOIN Facilities f ON b.facid = f.facid
GROUP BY f.facid
HAVING total_revenue < 1000
ORDER BY total_revenue;
"""

df = pd.read_sql_query(query, conn)
df
```

 Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order

```python
# Define the query
query = """
SELECT
    m.surname AS member_surname,
    m.firstname AS member_firstname,
    r.surname AS recommender_surname,
    r.firstname AS recommender_firstname
FROM Members m
LEFT JOIN Members r ON m.recommendedby = r.memid
ORDER BY m.surname, m.firstname;
"""

# Execute the query and show results
df = pd.read_sql_query(query, conn)
df
```

 Q12: Find the facilities with their usage by member, but not guests

```python
# Define query
query = """
SELECT
    f.facid,
    f.name AS facility_name,
    COUNT(b.bookid) AS usage_count
FROM Bookings b
JOIN Facilities f ON b.facid = f.facid
WHERE b.memid <> 0
GROUP BY f.facid, f.name
ORDER BY f.facid;
"""

# Execute query
df = pd.read_sql_query(query, conn)

df
```
